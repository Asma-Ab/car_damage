```python
import os
import csv

# Function to get all image paths and their corresponding labels
def get_image_paths_and_labels(folder_path):
    image_paths_labels = []
    for label in os.listdir(folder_path):
        label_path = os.path.join(folder_path, label)
        if os.path.isdir(label_path):
            for image_name in os.listdir(label_path):
                image_path = os.path.join(label_path, image_name)
                image_paths_labels.append((image_path, label))
    return image_paths_labels

# Function to write CSV file
def write_csv(csv_path, data):
    with open(csv_path, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["path", "label"])
        writer.writerows(data)

# Paths to the main dataset folders
dataset_path = "/content/drive/MyDrive/damage_severity_dataset/"

# Output CSV file paths
train_path = "/content/drive/MyDrive/damage_severity/train.csv"
valid_path = "/content/drive/MyDrive/damage_severity/valid.csv"
test_path = "/content/drive/MyDrive/damage_severity/test.csv"

# Get image paths and labels for train, valid, and test sets
train_data = get_image_paths_and_labels(os.path.join(dataset_path, "train"))
valid_data = get_image_paths_and_labels(os.path.join(dataset_path, "valid"))
test_data = get_image_paths_and_labels(os.path.join(dataset_path, "test"))

# Write CSV files
write_csv(train_path, train_data)
write_csv(valid_path, valid_data)
write_csv(test_path, test_data)
```
